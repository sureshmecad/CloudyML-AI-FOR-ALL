## Probability
https://youtu.be/uzkc-qNVoOk

The meaning of probability is basically the extent to which something is likely to happen. It is a branch of mathematics that deals with the occurrence of a random event. The value is always expressed from zero to one. 

This is the basic probability theory, which is also used in the probability distribution, to find the probability of a single event to occur, first, we should know the total number of possible outcomes.

### Formula for Probability
The probability formula is defined as the ratio of the number of favourable outcomes and the total number of outcomes.

**Probability of event to happen P(E) = Number of favourable outcomes/Total Number of outcomes**

**Example**: In a coin toss, what's the probability of getting head?

**Solution**: Total number of outcomes are 2 (head or tail)

what we want is head, so favorable outcomes is 1.

Therefore, 
                  
      P(E) = Number of favourable outcomes/Total Number of outcomes          
           = 1 / 2                    
           = 0.5

## Probability Distribution and Random Variables
https://youtu.be/cqK3uRoPtk0

https://youtu.be/YXLVjCKVP7U

Probability distribution is used to describe the possible outcomes for a random event. It is defined based on the sample space which is a set of possible outcomes of a random experiment (whose outcome cannot be predicted). Example, if we toss a coin, we cannot predict the outcome as either it will come as Head or Tail. The possible result of a random experiment is called an outcome. And the set of outcomes is called a sample point. We create a probability pattern table in terms of variable and probabilities with the help of these experiments or events.

### Random Variable
In a random experiment, the outcome can be a concrete object— a person, a lot of a manufactured item, a vehicle on the road. These objects will possess some qualities that are measurable— height or weight of a person, the number of defective components in the lot, the amount pulp that you can get from the mango, the number of people on the scooter. We want to model these measurable attributes, i.e., we would like to deal with numbers associated with the objects, rather than the objects themselves.

Example, On a coin toss, say outcome is 1 if it is a Head and 0 if it is a Tail. Alternately, map outcome of Head to +1 and Tail to -1. In all of the above, we have converted outcomes of a random experiment to numbers (integers or real), i.e., we have mapped an outcome of an experiment to a numerical value. Therefore we can say, the random variable is the numerical value of the outcome of a random experiment.



###Types of Probability Distribution Functions

A random variable has a probability distribution, which defines the probability of its unknown values. Random variables can be discrete (not constant) or continuous or both. So, there are two types of probability distribution which are used for different purposes.
1. Discrete Probability Distributions
2. Continuous Probability Distributions

### Discrete Probability Distributions
If the random variable associated with the probability distribution is discrete, then such a probability distribution is called discrete. Such a distribution is specified by a probability mass function (p).

Example: Bernoulli distribution, Binomial distribution. 

### Continuous Probability Distributions
If the random variable associated with the probability distribution is continuous, then such a probability distribution is said to be continuous. Such a distribution is specified by a probability mass function (ƒ).

Example: Normal or Gaussian distribution.




### PDF (Probability Density Function)

PDF is a statistical term that describes the probability distribution of the continuous random variable. Example: Height distribution of men in India.
 
The formula for PDF (fX(x)):

Prob(x < X ≤ x + ∆x) ≈ fX(x)∆x 



### PMF (Probability Mass Function):-

pX(k) = Prob(X = k)

PMF is a statistical term that describes the probability distribution of the discrete random variable. For e.g, Throwing a dice (You can only select 1 to 6 numbers (countable) )

### CDF (Cumulative Distribution Function):-
Fx(x) = Prob(X ≤ x)

Fx(x) is called the Cumulative Distribution Function (CDF) of X.

PMF is only applicable for discrete random variables and PDF is only applicable for continuous random variables. That's why we have cumulative distribution function as it is applicable for describing the distribution of random variables either it is continuous or discrete.

For example, if X is the height of a person selected at random then Fx(x) is the chance that the person will be shorter than x. If F(175 cm)=0.8. then there is an 80% chance that a person selected at random will be shorter than 180 cm (equivalently, a 20% chance that they will be taller than 175cm)

### Properties of CDF for Continuous Spectrum

Fx(−∞) = 0.

Fx(∞) = 1.

If x1 ≤ x2, then Fx(x2) ≥ Fx(x1)

Follows because

{X ≤ x1} ⊆ {X ≤ x2}


This means that Fx(x) is a non decreasing function
Prob(x1 < X ≤ x2) = Fx(x2) − Fx(x1).

Prob(X > x) = 1 − Fx(x)

For continuous spectrum, PDF is regarded as derivative of CDF.

### Expectation and Variance

The expectation is defined as the weighted average value of a random variable.

The variance is defined as the average value of the square distance from the mean value of a random variable. It represents the distribution of random variable near the mean. Small variance indicates that the random variable is distributed near the mean value and Big variance indicates that the random variable is distributed far from the mean value.

              
              σ^2 = Var ( X ) = E(X^2) - μ^2

### Properties of Expected Values and Variance

- When X and Y are two independent random variables: E(XY) = E(X)E(Y)
- E(aX+bY) = aE(X) + bE(Y)
- Var(X+Y) = Var(X) + Var(Y) + 2Cov(X,Y) where Cov(X,Y) = E(XY) - E(X)E(Y).
- Var(constant) = 0 and E(constant) = constant
- Var(aX) = (a^2).Var(X)

## Bernoulli Distribution
https://youtu.be/nl9WiZMZnYs

Bernoulli random variable takes values in set {0, 1}.

X ∈ {0, 1}

Prob(X = 1) = pX(1) = p

Prob(X = 0) = pX(0) = 1 − p

Best visualised as a coin toss with head mapped to 1 and tail to 0.

The mean of Bernoulli random variable(X) is \\
E[X] = 1(p) +0(1-p) = p 

The variance of Bernoulli random variable is \\
V[X] = E[X²]-[E(X)]² = 1²p+0²(1-p)-p²=p(1-p)

##Binomial Distribution
https://youtu.be/_FbZI9mtSSM

In binomial probability distribution, we take sum of N independent Bernouilli random variables in our consideration, e.g., number of Heads from N independent coin tosses equivalently, count the number of heads in N tosses of a coin with bias p. 

The binomial distribution formula is for any random variable X, given by

P(x:n,p) = nCx.p^x.(1-p)^n-x
 
 Or

P(x:n,p) = nCx.p^x.(q)^n-x

Where,

n = the number of experiments

x = 0, 1, 2, 3, 4, …

p = Probability of Success in a single experiment

q = Probability of Failure in a single experiment = 1 – p

### Mean and Variance of a Binomial Distribution
Mean(µ) = np

Variance(σ**2) = npq

The variance of a Binomial Variable is always less than its mean [npq < np].

##Central Limit Theorem 
https://youtu.be/JNm3M9cqWyc

###Central Limit Theorem Statement
The central limit theorem states that whenever a random sample of size n is taken from any distribution with mean and variance, then the sample mean will be approximately normally distributed with mean and variance. The larger the value of the sample size, the better the approximation to the normal.

As the sample size gets bigger, the mean of the sample will get closer to the actual population mean. If the sample size is small, the actual distribution of the data may or may not be normal, but as the sample size gets bigger, it can be approximated by a normal distribution.

As you can see it from here as well that using central limit theorem, when the large sample size is having a finite variance, the samples will be normally distributed and the mean of samples will be approximately equal to the mean of the whole population or mathematically, any population with mean and standard deviation, the distribution of the sample mean for sample size N has mean μ and standard deviation σ / √n .

## Confidence Interval
https://youtu.be/czdwHU27OqA

The confidence level is actually the proportion/frequency of acceptable confidence intervals that contain the true value of any unknown parameter/our random variable. Mathematically, we calculate the confidence intervals using the given confidence level from an endless number of independent samples so that the proportion of the range contains the true value of the parameter that will be equal to the confidence level. The confidence interval is based on the mean and standard deviation. Mostly, the confidence level is selected before examining the data but the commonly used confidence level are 95%, 90% and 99% confidence level.  

### Confidence Interval Formula

Confidence Interval  = [lower bound, upper bound]

The formula to find CI is

X̄ ± Zα/2 × [ σ / √n ]

Where

X̄ = Mean

Z = Confidence coefficient

α = Confidence level

σ = Standard deviation

N = sample space

The value after the ± symbol is known as the margin of error.

## Hypotheses Testing
1. https://youtu.be/5D1gV37bKXY
2. https://youtu.be/Q1yu6TQZ79w
3. https://youtu.be/pfrZYfcfJEA

We run a hypothesis test that helps statisticians determine if the evidence are enough in a sample data to conclude that a research condition is true or false for the entire population. Here, we are not interested in explicitly estimating the parameters but in verifying a statistical hypotheses about them.

- value of single parameter: (1) μ = 1, (2) μ > 1
- values of a set of parameters: (1) μ = 10, σ > 1
- distribution of the population: uniform, normal, exponential, etc.

It is a hypothesis because it is not known if it is true. Test method is as follows: Design a suitable random experiment and accept the hypothesis if the data from the experiment is consistent with the hypothesis.
The notion of consistent needs to be made more precise. Accepting the hypothesis does not mean “claiming that the hypothesis is true; rather, we are saying experimental data appears to be consistent with the hypothesis. Once again notion of ‘appears to be’ will be made more precise.

Null Hypothesis, denoted by H0 is to state that the initial claim or prior
belief is true. 
- If there is the claim of a better drug, H0 will be to state that this is not better than the current drug. (cures a higher fraction, or has lesser side effects).
- If a new vendor to procure components is being tried, the new vendor is not any better than the current. (rejection ratio of the components is not lower).

The word null means “of no value, effect or consequence,” which
suggests that H0 should be identified with the hypothesis of no change,
no effect or no improvement.

### Test Procedures

- Conduct the random experiment and obtain the samples.
- Define the Test Statistic, a function of the sample data on which the
decision to accept or reject H0 is based.
- Define a Rejection Region, the set of all test statistic values for which
H0 will be rejected.
- We need to see what are the possible errors from our experiment and conclusion. 
- Type I error: Rejecting H0 when H0 is indeed true. The probability of Type I error is usually denoted by α.
- Type II error: Not rejecting H0 when H0 is false. The probability of Type II error is usually denoted by β.
